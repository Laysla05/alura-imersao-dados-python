```python
#coluna traduzida
columns_translation = {
    'ano': 'ano',
    'nivel_experiencia': 'experiencia',
    'tipo_emprego': 'emprego',
    'cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_usd': 'usd',
    'residencia_funcionario': 'residencia',
    'taxa_remoto': 'remoto',
    'localizacao_empresa': 'empresa',
    'tamanho_empresa': 'tamanho_empresa'
}
df = df.rename(columns=columns_translation)
print(df.columns)
```

```python
df["experiencia"].value_counts()
```

```python
df["emprego"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
experience_translation = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Diretor'
}
df['experiencia'] = df['experiencia'].replace(experience_translation)

print(df['experiencia'].value_counts())
```

```python
employment_type_translation = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}
df['emprego'] = df['emprego'].replace(employment_type_translation)

print(df['emprego'].value_counts())
```

```python
remote_translation = {
    0: 'Presencial',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(remote_translation)

print(df['remoto'].value_counts())
```
